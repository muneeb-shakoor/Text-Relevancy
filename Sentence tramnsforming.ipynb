```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2') #stsb-roberta-large
```

```python
import pandas as pd
data = pd.read_excel('/content/gdrive/My Drive/NLP-Inaequo/DataPoints.xlsx', index_col=None)
data['Embedding Score']= np.nan
data.head()
```

```python
i = 0
while i < (len(data.index)-2):
  embedding1 = model.encode(data['Original'][i], convert_to_tensor=True)
  for j in range(3):
    embedding2 = model.encode(data['Answers'][i+j], convert_to_tensor=True)
    cosine_score = util.pytorch_cos_sim(embedding1, embedding2)
    data['Embedding Score'][i+j] = cosine_score[0][0].item()
  i = i+3
data.head
```

```python
data.to_excel("/content/gdrive/My Drive/NLP-Inaequo/Embedding_scores.xlsx") 
```
